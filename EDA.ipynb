```python
# Basic statistical characterstics
df.describe()
```

```python
# In order to see statistics of non numerical features
df.describe(include=['object', 'bool'])
```

```python
#For categorical (type object) and boolean (type bool) features we can use the value_counts method
df['name'].value_counts()
```

```python
# you can see other columns as well
df['id'].value_counts()
```

```python
# To calculate fractions
df['name'].value_counts(normalize=True)
```

# Sorting

```python
df.sort_values(by='price', ascending=False).head()
```

```python
# Sorting with more than 1 column
df.sort_values(by=['price', 'minimum_nights'],
        ascending=[False, False]).head()
```

```python
# Finding mean propotion of prices (y variable)
df['price'].mean()
```

```python
# Average of all stats for a particular neighbourhood group
df[df['neighbourhood_group'] == "Manhattan"].mean()
```

```python
# Average of price for a particular neighbourhood group
df[df['neighbourhood_group'] == "Manhattan"]['price'].mean()
```

```python
# Average price of manhattan rooms with more than 5 minimum nights
df[(df['neighbourhood_group'] == "Manhattan") & (df['minimum_nights'] > 5)]['price'].mean()
```

##DataFrames can be indexed by column name (label) or row name (index) or by the serial number of a row. The loc method is used for indexing by name, while iloc() is used for indexing by number.

```python
# Last line
df[-1:]
```

```python
#Lambda functions
#All neighbourhood groups that start with M
df[df['neighbourhood_group'].apply(lambda neighbourhood_group: neighbourhood_group[0] == 'M')].head()
```

```python
#The map method can be used to replace values in a column by passing a dictionary of the form {old_value: new_value} as its argument
'''
d = {'No' : False, 'Yes' : True}
df['International plan'] = df['International plan'].map(d)
df.head()

'''
```

```python
#The same thing can be done with the replace method:
d = {'Entire home/apt' : 'Full House'}

df = df.replace({'room_type': d})
df.head()
```

## Group By

#df.groupby(by=grouping_columns)[columns_to_show].function()

```python
columns_to_show = ['neighbourhood_group', 'room_type']

df.groupby(['price'])[columns_to_show].describe(percentiles=[])
```

```python
columns_to_show = ['neighbourhood_group', 'price']

df.groupby(['room_type'])[columns_to_show].agg([np.mean, np.std, np.min,np.max])
```

# Summary Tables

```python
# Getting a cross tab of items
pd.crosstab(df['neighbourhood_group'], df['room_type'])
```

```python
# Getting a cross tab of items in percentage share
pd.crosstab(df['neighbourhood_group'], df['room_type'], normalize = True)
```
